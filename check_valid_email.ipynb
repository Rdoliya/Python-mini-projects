```python
# check email address validation
email = input("Enter your email address: ")
k, j, d = 0, 0, 0

if len(email) >= 6:  # condition 1
    if email[0].isalpha():  # condition 2
        if "@" in email and email.count("@") == 1:  # condition 3
            if email[-4] == "." or email[-3] == ".":  # condition 4
                for i in email:
                    if i.isspace():  # condition 5 - fixed
                        k = 1
                    elif i.isupper():  # condition 5 - fixed
                        j = 1
                    elif i.isdigit():  # condition 5
                        continue
                    elif i == "_" or i == "." or i == "@":  # condition 5
                        continue
                    elif i.isalpha():  # condition 5
                        continue
                    else:
                        d = 1
                if k == 1 or j == 1 or d == 1:
                    print("Wrong email 5")
                else:
                    print("Valid email")
            else:
                print("wrong email 4")
        else:
            print("wrong email 3")
    else:
        print("wrong email 2")
else:
    print("wrong email 1")
```
