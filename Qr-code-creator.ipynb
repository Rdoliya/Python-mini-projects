```python
#create a QR code for any URL and save it as "qr.png"
import qrcode as qr
img = qr.make("https://www.linkedin.com/in/rishyup-doliya-833b352b0/")
img.save("qr.png")
```

```python
#change the color of the QR code to red and the background to white
import qrcode
from PIL import Image
qr = qrcode.QRCode(version = 1, error_correction = qrcode.constants.ERROR_CORRECT_H, box_size = 10, border = 4)
qr.add_data("https://www.linkedin.com/in/rishyup-doliya-833b352b0/")
qr.make(fit = True)
img = qr.make_image(fill_color = "red", back_color = "white")
img.save("qr.png")
```
